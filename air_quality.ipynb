```python
import pandas as pd
import matplotlib.pyplot as plt
file_name = "data.csv" #load database
data = pd.read_csv(file_name) #read data
```

```python
#########Single day analysis##########

id="a8-17-58-ff-fe-03-0d-51" #enter sensor  id for testing different sensor 
day="2017-06-27" #enter day for testingdifferent day 
attribute="light" #enter attribute ex- co2
value=[]
time=[]
morning_time=[]
morning_value=[]
lunch_time=[]
lunch_value=[]
ev_time=[]
ev_value=[]
night_time=[]
night_value=[]
for i in range(34999,102999):#this range select data from 12 am to 12 pm single day
    if data["id"][i]==id:
        if data["timestamp"][i].split(" ")[0].replace(" ", "")==day:
            time_all=data["timestamp"][i].split(" ")[1].replace(" ", "").split(".")[0]
            hour=time_all.split(":")[0].replace(" ", "")#split it hours
            if int(hour)>=8 and int(hour)<12:#start from 8am
                morning_time.append(time_all)
                morning_value.append(data[attribute][i])
            elif int(hour)>=12 and int(hour)<15:
                lunch_time.append(time_all)
                lunch_value.append(data[attribute][i])
            elif int(hour)>=15 and int(hour)<19:
                ev_time.append(time_all)
                ev_value.append(data[attribute][i])
            elif int(hour)>=19 and int(hour)<=23:
                night_time.append(time_all)
                night_value.append(data[attribute][i])

```

```python
print("Single day Measurement of "+attribute+" for sensor id : "+id+" at day "+day)
print("Time duration Morning 8 am -12 pm:")
print("Max Value of "+attribute+" :"+str(max(morning_value)))
print("Min Value of "+attribute+" :"+str(min(morning_value)))
print("Difference of  "+attribute+" :"+str(max(morning_value)-min(morning_value)))


print("Time duration Lunch hour 12pm -3pm:")
print("Max Value of "+attribute+" :"+str(max(lunch_value)))
print("Min Value of "+attribute+" :"+str(min(lunch_value)))
print("Difference of  "+attribute+" :"+str(max(lunch_value)-min(lunch_value)))


print("Time duration Evening hour 3pm -7pm:")
print("Max Value of "+attribute+" :"+str(max(ev_value)))
print("Min Value of "+attribute+" :"+str(min(ev_value)))
print("Difference of  "+attribute+" :"+str(max(ev_value)-min(ev_value)))


print("Time duration night hour 7pm -12pm:")
print("Max Value of "+attribute+" :"+str(max(night_value)))
print("Min Value of "+attribute+" :"+str(min(night_value)))
print("Difference of  "+attribute+" :"+str(max(night_value)-min(night_value)))
```

```python
########plot for signle day analysis##########
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl

def demo(sty):
    mpl.style.use(sty)
    fig, ax = plt.subplots()

    ax.set_title('style: {!r}'.format(sty), color='C0')

    ax.plot(morning_time, morning_value, 'C1', label='Morning 8 am -12 pm')
    ax.plot(lunch_time, lunch_value, 'C2',label='Lunch 12 -3pm')
    ax.plot(ev_time, ev_value, 'C3', label='Evening 3-7 pm')
    ax.plot(night_time, night_value, 'C4',label='Night 7pm-12am')

    ax.legend()

demo('default')
```

```python
#weekly  variations analysis

id="a8-17-58-ff-fe-03-0d-51" #enter sensor  id for testing different sensor 
day=["2017-07-01", "2017-07-02","2017-07-03","2017-07-04","2017-07-05","2017-07-06","2017-07-07",] #enter day for testingdifferent day 
attribute="temperature" #enter attribute ex- co2
value=[]
time=[]
first_day_time=[]
first_day_value=[]
second_day_time=[]
second_day_value=[]
third_day_time=[]
third_day_value=[]
fourth_day_time=[]
fourth_day_value=[]
fifth_day_time=[]
fifth_day_value=[]
sixth_day_time=[]
sixth_day_value=[]
seventh_day_time=[]
seventh_day_value=[]

for i in range(0,99999):#this range select data for 7 day
    if data["id"][i]==id:       
        if data["timestamp"][i].split(" ")[0].replace(" ", "")==day[0]:
            time_all=data["timestamp"][i].split(" ")[1].replace(" ", "").split(".")[0]
            first_day_time.append(time_all)
            first_day_value.append(data[attribute][i])
        elif data["timestamp"][i].split(" ")[0].replace(" ", "")==day[1]:
            time_all=data["timestamp"][i].split(" ")[1].replace(" ", "").split(".")[0]
            second_day_time.append(time_all)
            second_day_value.append(data[attribute][i])

        elif data["timestamp"][i].split(" ")[0].replace(" ", "")==day[2]:
            time_all=data["timestamp"][i].split(" ")[1].replace(" ", "").split(".")[0]
            third_day_time.append(time_all)
            third_day_value.append(data[attribute][i])

        elif data["timestamp"][i].split(" ")[0].replace(" ", "")==day[3]:
            
            time_all=data["timestamp"][i].split(" ")[1].replace(" ", "").split(".")[0]
            fourth_day_time.append(time_all)
            fourth_day_value.append(data[attribute][i])

        elif data["timestamp"][i].split(" ")[0].replace(" ", "")==day[4]:
            time_all=data["timestamp"][i].split(" ")[1].replace(" ", "").split(".")[0]
            fifth_day_time.append(time_all)
            fifth_day_value.append(data[attribute][i])

        elif data["timestamp"][i].split(" ")[0].replace(" ", "")==day[5]:
            time_all=data["timestamp"][i].split(" ")[1].replace(" ", "").split(".")[0]
            sixth_day_time.append(time_all)
            sixth_day_value.append(data[attribute][i])

        elif data["timestamp"][i].split(" ")[0].replace(" ", "")==day[6]:
            time_all=data["timestamp"][i].split(" ")[1].replace(" ", "").split(".")[0]
            seventh_day_time.append(time_all)
            seventh_day_value.append(data[attribute][i])

```

```python
########print weekly data
print("Weely Measurement of "+attribute+" for sensor id : "+id+" from day "+day[0]+" to "+day[6])
print("First Day:")
print("Max Value of "+attribute+" :"+str(max(first_day_value)))
print("Min Value of "+attribute+" :"+str(min(first_day_value)))
print("Difference of  "+attribute+" :"+str(max(first_day_value)-min(first_day_value)))

print("Second Day:")
print("Max Value of "+attribute+" :"+str(max(second_day_value)))
print("Min Value of "+attribute+" :"+str(min(second_day_value)))
print("Difference of  "+attribute+" :"+str(max(second_day_value)-min(second_day_value)))

print("Third Day:")
print("Max Value of "+attribute+" :"+str(max(third_day_value)))
print("Min Value of "+attribute+" :"+str(min(third_day_value)))
print("Difference of  "+attribute+" :"+str(max(third_day_value)-min(third_day_value)))

print("Fourth Day:")
print("Max Value of "+attribute+" :"+str(max(fourth_day_value)))
print("Min Value of "+attribute+" :"+str(min(fourth_day_value)))
print("Difference of  "+attribute+" :"+str(max(fourth_day_value)-min(fourth_day_value)))

print("Fifth Day:")
print("Max Value of "+attribute+" :"+str(max(fifth_day_value)))
print("Min Value of "+attribute+" :"+str(min(fifth_day_value)))
print("Difference of  "+attribute+" :"+str(max(fifth_day_value)-min(fifth_day_value)))

print("Sixth Day:")
print("Max Value of "+attribute+" :"+str(max(sixth_day_value)))
print("Min Value of "+attribute+" :"+str(min(sixth_day_value)))
print("Difference of  "+attribute+" :"+str(max(sixth_day_value)-min(sixth_day_value)))

print("Seventh Day:")
print("Max Value of "+attribute+" :"+str(max(seventh_day_value)))
print("Min Value of "+attribute+" :"+str(min(seventh_day_value)))
print("Difference of  "+attribute+" :"+str(max(seventh_day_value)-min(seventh_day_value)))
```

```python
#plot for weekly analysis
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl

def demo(sty):
    mpl.style.use(sty)
    fig, ax = plt.subplots()
    ax.set_title('style: {!r}'.format(sty), color='C0')
    ax.plot(first_day_time, first_day_value, 'C1', label='Saturday')
    ax.plot(second_day_time, second_day_value, 'C2',label='Sunday')
    ax.plot(third_day_time, third_day_value, 'C3', label='Monday')
    ax.plot(fourth_day_time, fourth_day_value, 'C4',label='Tuesday')
    ax.plot(fifth_day_time, fifth_day_value, 'C5', label='Wednesday')
    ax.plot(sixth_day_time, sixth_day_value, 'C6',label='Thrusday')
    ax.plot(seventh_day_time, seventh_day_value, 'C7', label='Friday')

    ax.legend()

demo('default')
```

```python
#monthly avarage data calculation 3 month 
import pandas as pd
start_date = '2017-06-27'
end_date = '2017-09-27'
attribute="humidity"
dates=[]
sensor1=[]
sensor2=[]
sensor3=[]
sensor4=[]
sensor5=[]
sensor6=[]
sensor7=[]

id=["a8-17-58-ff-fe-03-0d-51","a8-17-58-ff-fe-03-0d-4d","a8-17-58-ff-fe-03-0d-50","a8-17-58-ff-fe-03-0d-4f","a8-17-58-ff-fe-03-0d-49","a8-17-58-ff-fe-03-0d-4c","a8-17-58-ff-fe-03-0d-4b"]
mydates = pd.date_range(start_date, end_date).tolist()
for i in range(len(mydates)):
    dates.append(str(mydates[i]).split(" ")[0].replace(" ", "")) #make a list of all date in the given interval

for i in range(0,1700000,10):#this range select data from June to september
    if data["timestamp"][i].split(" ")[0].replace(" ", "") in dates:
        if data["id"][i]==id[0]:
            sensor1.append(data[attribute][i])
        if data["id"][i]==id[1]:
            sensor2.append(data[attribute][i])
        if data["id"][i]==id[2]:
            sensor3.append(data[attribute][i])
        if data["id"][i]==id[3]:
            sensor4.append(data[attribute][i])
        if data["id"][i]==id[4]:
            sensor5.append(data[attribute][i])
        if data["id"][i]==id[5]:
            sensor6.append(data[attribute][i])
        if data["id"][i]==id[6]:
            sensor7.append(data[attribute][i])
```

```python
########print monthly  data
import numpy as np
sensors=[sensor1,sensor2,sensor3,sensor4,sensor5,sensor6,sensor7]
print("Monthly measurement of "+attribute+" from day "+start_date+" to "+end_date)
for i in range (1,8):
    print("measurement of "+attribute+" for "+"sensor"+str(i)+", sensor id : "+" "+id[i-1])
    print("Max Value of "+attribute+" : "+str(max(sensors[i-1])))
    print("Min Value of "+attribute+" :"+str(min(sensors[i-1])))
    print("Average of "+attribute+" :"+str(np.average(sensors[i-1])))
    
```

```python
#plot for 3 monthly average analysis
import numpy as np
import matplotlib.pyplot as plt

# define some sizes of the scatter marker

fig, ax = plt.subplots()

ax.plot(["sensor1","sensor1"], [1,11],  'bo', label='Sensor1')
ax.scatter("sensor2", np.average(sensor2), facecolor='red', label='Sensor2')
ax.scatter("sensor3", np.average(sensor3), facecolor='green', label='Sensor3')
ax.scatter("sensor4", np.average(sensor4), facecolor='yellow', label='Sensor4')
ax.scatter("sensor5", np.average(sensor5), facecolor='black', label='Sensor5')
ax.scatter("sensor6", np.average(sensor6), facecolor='gray', label='Sensor6')
ax.scatter("sensor7", np.average(sensor7), facecolor='pink', label='Sensor7')
plt.legend(('s1', 's2', 's3','s4','s5','s6','s7'))

plt.show()
```
